Loading Data

```python
import pandas as pd
import numpy as np
import nltk

df = pd.read_excel("C:/Users/Mirtha/Downloads/Source Code Skripsi/User Story and User Story Scenario_Revisi 2.xlsx")
df = df.fillna('')
df
```

Menggabungkan Data

```python
# Menggabungkan data setiap kolom
df['combined'] = df.iloc[:, :13].apply(lambda x: '\n'.join(x), axis=1)

# Menampilkan hasil gabungan setiap kolom
print(df.loc[0, 'combined'])
```

Numeric Feature Extraction

```python
# Menghitung jumlah token dan skenario
df['combined'] = df['combined'].str.lower()
df['number_word'] = df['combined'].apply(lambda x: len(x.split()))
df['number_scenario'] = df['combined'].apply(lambda x: x.count("scenario:"))
print(df['number_word'])
print(df['number_scenario'])

# Normalisasi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print("Hasil normalisasi")
df[['number_word', 'number_scenario']] = scaler.fit_transform(df[['number_word', 'number_scenario']])
df[['number_word', 'number_scenario']]
```

Text Feature Extraction

```python
# Text Preprocessing

# Case folding
print("Hasil case folding")
df['combined'] = df['combined'].str.lower()
print(df['combined'])

# Pembersihan teks dari tanda baca
import string
def remove_punctuation(text):
    punctuation_to_replace = string.punctuation + "‘’“”"
    translation_table = str.maketrans(punctuation_to_replace, ' ' * len(punctuation_to_replace))
    cleaned_text = text.translate(translation_table)
    return cleaned_text

df['combined'] = df['combined'].apply(remove_punctuation)
print("Hasil pembersihan teks dari tanda baca")
print(df.loc[0, 'combined'])

# Tokenisasi
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

df['combined'] = df['combined'].apply(lambda x: word_tokenize(x))
print("Hasil tokenisasi")
print(df['combined'])

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Stopword removal
stop_words = set(stopwords.words('english'))
def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in stop_words]

df['combined'] = df['combined'].apply(remove_stopwords)
print("Hasil stopword removal")
print(df['combined'])

# Lematisasi
lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

df['combined'] = df['combined'].apply(lemmatize_tokens)
print("Hasil lematisasi")
df['combined']
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Mengembalikan ke bentuk string
df['combined'] = df['combined'].apply(lambda x: ' '.join(x))

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined'])

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df
```

Data Integration

```python
# Concatenate 'effort' column of the original DataFrame with TF-IDF DataFrame
integrated_data = pd.concat([tfidf_df, df[['number_word', 'number_scenario', 'Effort']]], axis=1)
integrated_data
```

Model (Text)

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score


# Data teks dan effort
X = tfidf_df
y = df["Effort"].to_numpy()

# One-hot encode the target variable
enc = OneHotEncoder(sparse=False)
y_encoded = enc.fit_transform(y.reshape(-1, 1)) 

# Adaboost
clf = MultiOutputClassifier(AdaBoostClassifier(random_state=42))
scorer = make_scorer(accuracy_score) # Menggunakan akurasi untuk scoring
cv_scores = cross_val_score(clf, X, y_encoded, cv=5, scoring=scorer) # Perform 5-fold cross validation

# Akurasi Adaboost
print("Cross-validation scores dari Adaboost:", cv_scores)
print("Mean Cross-validation score dari Adaboost:", np.mean(cv_scores))

# Decision Tree
clf = DecisionTreeClassifier(random_state=42)
cv_scores = cross_val_score(clf, X, y_encoded, cv=5, scoring=scorer) # Perform 5-fold cross validation

# Akurasi Decision Tree
print("Cross-validation scores dari Decision Tree:", cv_scores)
print("Mean Cross-validation score dari Decision Tree:", np.mean(cv_scores))


# Multinomial Naive Bayes
clf = MultinomialNB()
cv_scores = cross_val_score(clf, X, np.argmax(y_encoded, axis=1), cv=5, scoring=scorer)# Perform 5-fold cross validation

# Akurasi Multinomial Naive Bayes
print("Cross-validation scores dari Multinomial Naive Bayes:", cv_scores)
print("Mean Cross-validation score dari Multinomial Naive Bayes:", np.mean(cv_scores))

# Adaboost
clf = MultiOutputClassifier(AdaBoostClassifier(random_state=42))
scorer = make_scorer(accuracy_score) # Menggunakan akurasi untuk scoring
cv_scores = cross_val_score(clf, X, y_encoded, cv=10, scoring=scorer) # Perform 10-fold cross validation

# Akurasi Adaboost
print("Cross-validation scores dari Adaboost:", cv_scores)
print("Mean Cross-validation score dari Adaboost:", np.mean(cv_scores))

# Decision Tree
clf = DecisionTreeClassifier(random_state=42)
cv_scores = cross_val_score(clf, X, y_encoded, cv=10, scoring=scorer) # Perform 10-fold cross validation

# Akurasi Decision Tree
print("Cross-validation scores dari Decision Tree:", cv_scores)
print("Mean Cross-validation score dari Decision Tree:", np.mean(cv_scores))


# Multinomial Naive Bayes
clf = MultinomialNB()
cv_scores = cross_val_score(clf, X, np.argmax(y_encoded, axis=1), cv=10, scoring=scorer)# Perform 10-fold cross validation

# Akurasi Multinomial Naive Bayes
print("Cross-validation scores dari Multinomial Naive Bayes:", cv_scores)
print("Mean Cross-validation score dari Multinomial Naive Bayes:", np.mean(cv_scores))

```
